```python
df = pd.read_excel('Vendas.xlsx')
```

```python
df.head()
```

```python
#Lucro total
lucro = df['Valor Final'].sum()
print(f'O lucro total foi de {lucro:.2f}')
```

```python
Lucro_por_loja = df[['ID Loja', 'Valor Final']].groupby('ID Loja').sum()
display(Lucro_por_loja)
```

```python
# Vendas por loja

Vendas_por_loja_produto = df.groupby(['ID Loja', 'Produto'])['Valor Final'].sum().reset_index()
display(Vendas_por_loja_produto)
```
